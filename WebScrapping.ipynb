```python
import requests 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
url = "https://pk.ign.com/portal-2/193152/feature/the-top-100-video-games-of-all-time"

r = requests.get("https://pk.ign.com/portal-2/193152/feature/the-top-100-video-games-of-all-time")
htmlContent = r.content
print(htmlContent)
```

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify)
```

```python
title = soup.title
print(type(title.string))
print(title.get_text())
```

```python
paras = soup.find_all('p')
print(paras)
```

```python
print(soup.find('p').get_text())
print(soup.get_text())
```

```python
titles = soup.find_all(['h2'])
print('List of top 100 Games:', *titles, sep='\n\n')
```

```python
images = soup.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    print(image['src']+'\n')
```
